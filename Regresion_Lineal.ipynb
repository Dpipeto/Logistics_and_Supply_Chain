## 2. *Descripción del ejercicio*
- - -
> El ejercicio consiste en utilizar un modelo de regresión lineal para explorar la relación entre la produccion y demanda de electricidad. Comienza con la importación de bibliotecas escenciales para el manejo de datos y la creacion de graficos tales como NumPy, Matplotlib y scikit-learn. >>import matplotlib.pyplot as plt: Se importa Matplotlib (específicamente el módulo pyplot) para crear gráficos. plt es el alias que se usa comúnmente para referirse a este módulo.

>-electricity_demand = np.array([35, 70, 45, 85, 55, 75, 65, 95, 25, 90, 110, 40, 120, 15, 100, 130, 80, 115, 60, 125, 85, 105, 80, 140, 130, 35, 70, 145, 85, 55, 75, 65, 95, 25, 90, 120, 40, 120, 15, 90, 130, 80, 115, 60, 125, 85, 135, 80, 140, 130, 35, 70, 45, 85, 55, 75, 65, 40, 25, 98, 110, 40, 120, 15, 100, 130, 100, 115, 70, 155, 85, 105, 95, 140, 130, 85, 70, 45, 145, 55, 65, 65, 95, 25, 78, 110, 40, 120, 45, 100, 130, 80, 135, 60, 125, 85, 105, 90, 140, 130])
>> Electricity_production: Se crea un arreglo NumPy que contiene los datos de la produccion de la electricidad insertando datos originales y aleatorios escritos por los estudiantes.
>> Electricity_demand: Se crea un arreglo NumPy que contiene los datos de la demanda de electricidad correspondientes a la produccion en electricity_production.
>- x=electricity_production.reshape(-1,1)
>- y=electricity_demand
>>- x=electricity_production.reshape(-1,1): Reorganiza el arreglo de produccion de energia en un arreglo de 2 dimensiones con una columna (esto es necesario porque scikit-learn espera que las características tengan 2 dimensiones). El -1 permite que NumPy ajuste el número de filas automáticamente.
>>- y=electricity_demand: Asigna la demanda de electricidad al vector Y, que será la variable objetivo en la regresión.

>>-train_test_split: Divide los datos en dos subconjuntos:
>>- x_train y y_train: El 80% de los datos que se usarán para entrenar el modelo.
>>- x_test y y_test: El 20% restante se usará para probar el modelo.
test_size=0.2: Indica que el 20% de los datos será usado para pruebas.
>>- random_state=42: Asegura que la división sea reproducible, es decir, cada vez que corras el código, obtendrás la misma división de datos.
>- model=LinearRegression()
>- model.fit(x_train,y_train)
>>- model = LinearRegression(): Crea un modelo de regresión lineal usando la clase
LinearRegression.
>>- model.fit(x_train, y_train): Entrena el modelo utilizando los datos de entrenamiento (x_train y y_train). El modelo aprende la relación entre la produccion y demanda de electricidad.
>- y_pred=model.predict(x_test)
>>- model.predict(x_test): Usa el modelo entrenado para predecir la demanda de electricidad (y_pred) en el conjunto de prueba (x_test). Estas son las predicciones que luego compararemos con los valores reales.
>- r2_score=model.score(x_test,y_test)
>- print("R2 Score:",r2_score)
>- coefficient=model.coef_[0]
>- print("Coefficient:",coefficient)
>- intercept=model.intercept_
>- print("Intercept:",intercept)
>>- model.score(x_test, y_test): Calcula el coeficiente de determinación R2R^2R2, que mide qué tan bien el modelo se ajusta a los datos de prueba. Un valor más cercano a 1 indica un mejor ajuste.
>>- model.coef_[0]: Devuelve el coeficiente de regresión (la pendiente de la recta ajustada).
>>- model.intercept_: Devuelve el intercepto del modelo, que es el valor de y cuando x es igual a 0.
>- plt.scatter(x_test,y_test,color='red')
>- plt.plot(x_test,y_pred,color='blue',linestyle = '-', linewidth= 3)
>- plt.xlabel('electricity_production')
>- plt.ylabel('electricity_demand')
>- plt.title('electricity production vs electricity demand')
>- plt.legend(['Actual','Predicted'])
>- plt.show()
>>- plt.scatter: Crea un gráfico de dispersión (puntos) con los valores de prueba reales (x_test vs y_test), dibujando los puntos en azul.
>>- plt.plot: Dibuja una línea roja para mostrar las predicciones del modelo (x_test vs y_pred).
>>- plt.xlabel y plt.ylabel: Añade etiquetas a los ejes del gráfico.
>>- plt.title: Añade un título al gráfico.
>>- plt.legend: Muestra la leyenda del gráfico para distinguir los puntos reales de los predichos.
>>- plt.show(): Muestra el gráfico final.
